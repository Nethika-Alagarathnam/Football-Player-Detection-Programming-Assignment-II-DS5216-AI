```python
import os
import cv2

VIDEO_FOLDER  = r'D:\MSc\DS 5216 - Artificial Intelligence\PAS02_DTS2430\Football'
FRAMES_FOLDER = r'D:\MSc\DS 5216 - Artificial Intelligence\PAS02_DTS2430\dataset\images'

EXTRACT_EVERY = 10

VIDEO_MAP = {
    'Football 01.mp4': 'football',
    'Football 02.mp4': 'football',
    'Football 03.mp4': 'football',
    'Football 04.mp4': 'football',
    'Football 05.mp4': 'football',
    'Football 06.mp4': 'football',
    'Football 07.mp4': 'football',
    'Football 08.mp4': 'football',
    'Football 09.mp4': 'football',
    'Football 10.mp4': 'football',
}

os.makedirs(FRAMES_FOLDER, exist_ok=True)

total_saved = 0

for filename, sport in VIDEO_MAP.items():

    video_path = os.path.join(VIDEO_FOLDER, filename)

    if not os.path.exists(video_path):
        print(f'SKIP (not found): {filename}')
        continue

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f'ERROR opening video: {filename}')
        continue

    frame_idx = 0
    saved = 0

    clip_name = filename.replace(' ', '_').replace('.mp4', '')

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_idx % EXTRACT_EVERY == 0:
            fname = f'{sport}_{clip_name}_frame{frame_idx:04d}.jpg'
            out = os.path.join(FRAMES_FOLDER, fname)
            cv2.imwrite(out, frame)
            saved += 1

        frame_idx += 1

    cap.release()

    total_saved += saved
    print(f'{filename:25s} → {saved} frames extracted')

print(f'\nTotal frames extracted - {total_saved}')

print('Next step is to label images using LabelImg')
```
